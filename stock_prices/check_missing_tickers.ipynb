```python
import pandas as pd
from pathlib import Path
from IPython.display import display, HTML
import os
```

```python
BASE = Path(os.getcwd()).parent
FOLDER = BASE / "PROF/prof_components_extracted"
OUTPUT = "all_stock_prices.csv"
```

```python
# Load the extracted prices
prices_T = pd.read_csv(OUTPUT, index_col="Ticker")

# Get all tickers from the folder
all_tickers = sorted([f.stem for f in FOLDER.glob("*.csv")])

# Find missing tickers
found = set(prices_T.index.to_list())
missing = sorted(set(all_tickers) - found)

print(f"Missing: {len(missing)} | Found: {len(found)} | Total: {len(all_tickers)}")

# Display as scrollable HTML table
missing_df = pd.DataFrame(missing, columns=["Ticker"])
missing_df.index += 1  # start index at 1

html_table = missing_df.to_html(escape=False)
html_scroll = f"""
<div style="max-height: 500px; overflow-y: auto; border: 1px solid #ccc;">
    {html_table}
</div>
"""

display(HTML(html_scroll))
```
