```python
import lseg.data as lseg
import os
from dotenv import load_dotenv
from IPython.display import display, HTML
import matplotlib.pyplot as plt

load_dotenv()
app_key = os.getenv("LSEG_APP_KEY")

lseg.open_session("desktop.workspace", app_key=app_key)

test = lseg.get_history(
    universe="EQNR.OL",
    interval="monthly",
    start="2004-01-01",
    end="2024-12-31",
)

html_table = test.to_html()
html_scroll = f"""
<div style="max-height: 500px; overflow-y: auto; border: 1px solid #ccc;">
    {html_table}
</div>
"""

display(HTML(html_scroll))

fig, ax = plt.subplots(figsize=(12, 4))

ax.plot(test.index, test["OFF_CLOSE"], color="#2980b9", linewidth=1.5)
ax.set_title("Monthly close price")
ax.set_xlabel("Date")
ax.set_ylabel("Price")
ax.grid(axis="y", linewidth=0.5, alpha=0.5)

plt.tight_layout()
plt.show()

lseg.close_session()
```
